```python
def get_OHAREC():
    filelist = os.listdir('/home/chuylife10/dentalrecommender/OHAREC/')
    os.chdir('/home/chuylife10/dentalrecommender/OHAREC/')
    dataframes = [pd.read_sas(file) for file in filelist]
    
    getdataframe = pd.concat(dataframes)

    
    oharec = getdataframe['OHAREC'].fillna(3)
    
    return oharec

y = get_OHAREC()
```

```python
combined = [overral_data_09,overral_data_11,overral_data_13,overral_data_15]

overral_data = pd.concat(combined)



#overral_data_15 = overral_data_15.fillna(-1)


columns_to_drop =['SEQN']

X = overral_data.drop(columns=columns_to_drop)
#X.to_csv('/home/chuylife10/dentalrecommender/ov.csv')

#merging dataframes

#overral_data_09.to_csv(r'/home/chuylife10/dentalrecommender/overraldata09.csv')
#overral_data_11.to_csv(r'/home/chuylife10/dentalrecommender/overraldata11.csv')
#overral_data_13.to_csv(r'/home/chuylife10/dentalrecommender/overraldata13.csv')
#overral_data_15.to_csv(r'/home/chuylife10/dentalrecommender/overraldata15.csv')
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Descision Tree Classifier
from sklearn import tree

clfDT = tree.DecisionTreeClassifier(max_depth=8,max_leaf_nodes=300,min_samples_split=100,min_samples_leaf=20)
clfDT.fit(X_train, y_train)
clfDT_ypred0 = clfDT.predict(X_test)
```

```python
# Create DOT data
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
# what are the descriptors we are using so we can develope a reccomendation
# 

dot_data = tree.export_graphviz(clfDT, feature_names=X_train.columns, class_names='1234',impurity=False, out_file=None)
graph = pydotplus.graph_from_dot_data(dot_data)

graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn import metrics
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import cohen_kappa_score

#do train data 
#training data as previous years up to before most recent year
#test data as recent year
#use only questionare data
#drop NA
#training use 2 diff training sets 2010-15 and 2000- 2010
#compare data if diff just use 20 to 15 (yr)
results = {}
results['accuracy'] = metrics.accuracy_score(y_test, clfDT_ypred0)
results['f1_score'] = metrics.f1_score(y_test, clfDT_ypred0,average='weighted')# average='binary' by default
results['Recall'] = metrics.recall_score(y_test, clfDT_ypred0,average='weighted')
results['Precision'] = metrics. precision_score(y_test, clfDT_ypred0,average='weighted')
results['cohen_kappa'] = metrics.cohen_kappa_score(y_test, clfDT_ypred0)

print(results)
```
